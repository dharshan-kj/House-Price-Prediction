```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_excel('house data.xlsx')
```

```python
data
```

```python
plt.figure(figsize = (7,7))
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(data.area, data.price, color = 'violet', marker = '^')
```

```python
data_1 = data.drop('price',axis=1)
data_1
```

```python
price = data.price
price
```

```python
reg = LinearRegression()
reg.fit(data_1, price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
